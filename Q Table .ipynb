```python
# Import packages. 

import gym
import numpy as np
```

```python
# Create our enviornment. 

env = gym.make('MountainCar-v0')

# Reset the env

env.reset()

print(env.observation_space.high)
print(env.observation_space.low)
print(env.action_space.n)

# Trying to break it into chunks. 

os_size = [20] * len(env.observation_space.high)
discrete_os_win_size = (env.observation_space.high - env.observation_space.low) / os_size
print(discrete_os_win_size)

# Set up the Q table. 

q_table = np.random.uniform(low = -2, high = 0, size = (os_size + [env.action_space.n]))
print(q_table.shape)

done = False 

# The while loop to run the game. 

while not done: 
  action = 2
  new_state, reward, done, _ = env.step(action)
  print(reward, new_state)
  env.render()

env.close()
```
